```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.colors as mc
import colorsys


def plot_times():
    for problem_size in problem_sizes:
        df_ = df[(df["problem_size"]==problem_size)]
        plt.plot(df_["thread_num"], df_["time"]["mean"], "o--", label=f"Grid size = {problem_size}")
        plt.errorbar(df_["thread_num"], df_["time"]["mean"], df_["time"]["std"], linestyle='None', zorder=3, elinewidth=1, capsize=3)

    plt.title(f"Time of execution")
    plt.xlabel("Processors")
    plt.ylabel("Time [s]")
    plt.legend()
    plt.tight_layout()
    plt.savefig(f"plot1.jpg", dpi=300)
    plt.show()
    plt.clf()

def plot_effectiveness():
    for problem_size in problem_sizes:
        df_ = df[(df["problem_size"]==problem_size)]
        plt.plot(df_["thread_num"], (df_["time"]["mean"].values[0] / df_["time"]["mean"]) / df_["thread_num"], "o--", label=f"Grid size = {problem_size}")

    df_ = df[(df["problem_size"]==problem_size)]
    plt.plot(df_["thread_num"], df_["thread_num"] / df_["thread_num"], "-", color="#FFAAAA", label=f"Ideal effectivenes")

    plt.title(f"Effectiveness by processors")
    plt.xlabel("Processors")
    plt.ylabel("Effectiveness")
    plt.legend()
    plt.tight_layout()
    plt.savefig(f"plot3.jpg", dpi=300)
    plt.show()
    plt.clf()
    
def plot_speedup():
    plt.plot(df["thread_num"], df["thread_num"], "-", color="#FFAAAA", label=f"Ideal speedup")

    for problem_size in problem_sizes:
        df_ = df[(df["problem_size"]==problem_size)]
        plt.plot(df_["thread_num"], df_["time"]["mean"].values[0] / df_["time"]["mean"], "o--", label=f"Grid size = {problem_size}")

    df_ = df[(df["problem_size"]==problem_size)]
    plt.title(f"Speedup by processor count")
    plt.xlabel("Processors")
    plt.ylabel("Speedup")
    plt.legend()
    plt.tight_layout()
    plt.savefig(f"plot2.jpg", dpi=300)
    plt.show()
    plt.clf()
```
